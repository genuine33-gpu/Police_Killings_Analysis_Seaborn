## 1a: Box plot of age in both genders separeted by manner of death.

```python
df.manner_of_death.unique()
```

#### INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
df.manner_of_death.dropna(inplace=True, axis=0)
df.manner_of_death.isna().sum()
```

```python
df.shape
```

```python
sns.boxplot(data=df, x="gender", y="age", hue="manner_of_death")

# Eksen etiketlerini ve başlığı ayarlayın
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title(" Police Killings  based on Gender and Age")

# Legend ekleyin
plt.legend(title="manner_of_death")

# Grafiği gösterin
plt.show()
```

## 1b: Swarm Plot

#### INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
sns.swarmplot(data=df, x="gender", y="age", hue="manner_of_death")

# Eksen etiketlerini ve başlığı ayarlayın
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title(" Police Killings  based on Gender and Age")

# Legend ekleyin
plt.legend(title="manner_of_death")

# Grafiği gösterin
plt.show()
```

## 2: Plot 12 most dangerous cities

```python
df.head(2)
```

```python
df.groupby("city")["manner_of_death"].count().nlargest(12)

```

```python
s_array= df.groupby("city")["manner_of_death"].count().nlargest(12)
s_array.values
```

```python
s_array.index
```

```python
plt.figure (figsize = (8,6))
sns.barplot(x=s_array.index, y= s_array.values )
plt.xticks(rotation = 45, color = "b")
plt.title ("MostDangerous Cities", color = "r");
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='city', order=s_array.index)

# Eksen etiketlerini ve başlığı ayarlayın
plt.xlabel("City")
plt.ylabel("Count")
plt.title("MostDangerous Cities", color ="b")

# X ekseni etiketlerini döndürme
plt.xticks(rotation=45, color="b");
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
s_array_state= df.groupby("state")["manner_of_death"].count().nlargest(20)
s_array_state
```

```python
plt.figure (figsize = (8,6))
sns.barplot(x=s_array_state.index, y= s_array_state.values )
plt.xticks(rotation = 45, color = "b")
plt.title ("MostDangerous States", color = "b");
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='state', order=s_array_state.index)

# Eksen etiketlerini ve başlığı ayarlayın
plt.xlabel("State")
plt.ylabel("Count")
plt.title("Most Dangerous States", color ="b")

# X ekseni etiketlerini döndürme
plt.xticks(rotation=45, color="b");
```
